```python
import json
import random
import requests
import time
import pandas as pd
import os
from time import sleep
from time import time
from random import randint
```

```python
def requests_web_data(url):
    try:
        headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"}
                   #,"Cookie": ""}
        r = requests.get(url, headers=headers)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
    except:
        print('requests error!')
    else:
        return r.content
```

```python
time_ids = [str(i) for i in range(586,295388,360)]
print(len(time_ids))
```

```python
#time_ids = [str(i) for i in range(586,295388,360)]

time_date =[]
for time_id in time_ids:
    time_id_url = 'https://www.eecso.com/test/weibo/apis/getlatest.php?timeid=' + str(time_id)
    time_data = json.loads(requests_web_data(time_id_url).decode('utf-8'))
    time_date.append(time_data)

weibo_hot_data = []
for time_id in time_ids:
    historical_data_url = 'https://www.eecso.com/test/weibo/apis/currentitems.php?timeid=' + str(time_id)
    print(time_id)
    sleep(randint(1,3))
    data = json.loads(requests_web_data(historical_data_url).decode('utf-8'))
    weibo_hot_data.append(data)
```

```python
resou = pd.DataFrame(columns=['number','time_id','title','上榜时间','最后时间','life_span'])

import datetime

# 看爬取的时间，time_id是一天取两个，就是12：00、24：00.

a = 0
for i in weibo_hot_data[:len(weibo_hot_data)]:    #设置最后一个time_id不要，因为会多出来一个
    a +=1
    for j in i:
        time1 = datetime.datetime.strptime(j[1],'%Y-%m-%d %H:%M:%S')
        time2 = datetime.datetime.strptime(j[2],'%Y-%m-%d %H:%M:%S')
        delta = time1-time2  #计算存活周期
        
        resou = resou.append({'number':a,
                              'time_id':time_ids[weibo_hot_data.index(i)],
                              'title':j[0],
                              '上榜时间':j[2],
                              '最后时间':j[1],
                              'life_span':delta.seconds},ignore_index=True)
resou.head(10)
```
